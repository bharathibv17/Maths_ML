```python
import numpy as np
```

### Dot Product of 2 Vectors

```python
a = np.array([2,3])

b = np.array([4,4])

a_dot_b = np.dot(a,b)# multiplication of two values 
```

```python
print(a_dot_b)
```

```python
c = np.array([40,20])

d = np.array([30,50])

c_dot_d = np.dot(c,d)
```

```python
print(c_dot_d)
```

```python
x = np.array([40,20,80])

y = np.array([30,50,10])

x_dot_y = np.dot(x,y)
```

```python
print(x_dot_y)
```

Here, we can see the result in scaler

### Cross Product of 2 Vectors

```python
a = np.array([2,3])

b = np.array([4,4])

a_cross_b = np.cross(a,b) # cross product of 2 vector
```

```python
print(a_cross_b)
```

```python
x = np.array([40,20,80])

y = np.array([30,50,10])

x_cross_y = np.cross(x,y)
```

```python
print(x_cross_y)
```

Here the result of cross product is vector

### Projection of 'a' Vector on 'v' vector

```python
a = np.array([2,5])

v = np.array([8,-6])

#magnitude of v vector
magnitude_of_v = np.sqrt(sum(v**2))

proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v
```

```python
print("Projection of 'a'vector on 'v' vector ",proj_of_a_on_v)
```
