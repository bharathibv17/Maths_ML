```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

### Plotting a Vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

This is our vector, here x axis is not in correct way

```python
plt.quiver(0,0,4,5, scale_units='xy',angles='xy',scale=1)
plt.xlim(-8,+8) #xlim -->Xlimit
plt.ylim(-8,+8) #ylim -->Ylimit
plt.show()
```

Here base points starts at it origin, and co-ordinate of vector is[4,5]

```python
plt.quiver(0,0,4,5, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-3,-6, scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,+8) #xlim -->Xlimit
plt.ylim(-8,+8) #ylim -->Ylimit
plt.show()
```

## Addition of two vectors

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

sum = vector_1 + vector_2

print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,5,1, scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,+8) #xlim -->Xlimit
plt.ylim(-8,+8) #ylim -->Ylimit
plt.show()
```

## Subtraction of two Vectors

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

difference = vector_1 - vector_2

print(difference)
```

```python
plt.quiver(0,0,2,3, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,-1,1, scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,+8) #xlim -->Xlimit
plt.ylim(-8,+8) #ylim -->Ylimit
plt.show()
```

## Multiplying two Vectors

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = 2*vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,4,6, scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,+8) #xlim -->Xlimit
plt.ylim(-8,+8) #ylim -->Ylimit
plt.show()
```

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = -0.5*vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-1,-1.5, scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,+8) #xlim -->Xlimit
plt.ylim(-8,+8) #ylim -->Ylimit
plt.show()
```
